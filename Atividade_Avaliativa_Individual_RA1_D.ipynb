```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np


def Tempo(Nucleos):
  T = 10/np.log2(Nucleos)

  return T

nNucleos = [2,4,8,16,32,64]

T = Tempo(nNucleos)

plt.plot(T,nNucleos,"o")
plt.xlabel("Tempo")
plt.ylabel("Núcleos")
```

```python
def funcaoA(x):
  return 2*3**x - 6**x - 48

x = np.linspace(-5,5,500)

y = funcaoA(x)

plt.plot(x, y)

plt.xlabel(r'$2\times3^{x} - 6^{x} - 48$')
plt.ylabel("Decrescente")
plt.title("Função A")
```

```python
def funcaoB(x):
  return np.log10(x**2 - 3*x) - np.log10(x - 3) - 1

np.linspace(-5,5,500)

a = 1
b = -13
c = 30

delta = b**2 - 4*a*c

x1 = (-b + np.sqrt(delta))/(2*a)
x2 = (-b - np.sqrt(delta))/(2*a)

x = np.linspace(x1-2, x2+2, 100)

y = funcaoB(x)

plt.plot(x, y)

plt.xlabel(r"$\log_{10}(x^2-3x) - \log_{10}(x-3) - 1$")
plt.ylabel("Crescente")
plt.title("Função B")
```

$\displaystyle \lim_{x->1} = \frac{(x^{10})-1}{x^5-1}

$\displaystyle \lim_{x->1} = (x^{10}-1) = (x^5+1)(x^5-1) \displaystyle$

$\displaystyle \lim_{x->1} = (x^5-1 = (x-1)(x^4 + x^3 + x^2 + x + 1)\displaystyle$

$\displaystyle \lim_{x->1} = \frac{x^{10}-1}{x^5-1} = \frac{[(x^5+1)(x^5-1)]}{[(x^4+x^3+x^2+x+1)]} \displaystyle$


$\displaystyle \lim_{x->1} = \frac{x^{10}-1}{x^5-1} = \frac{(x^5+1)}{[(x^4+x^3+x^2+x+1)]} \displaystyle$

$\displaystyle \lim_{x->1} = \frac{(1^5+1)}{[(1^4+1^3+1^2+1+1)]} = \frac{2}{5}\displaystyle$



```python
def funcA(x):
  return (x**10 - 1)/(x**5 - 1)

x =np.linspace(0.9,1.1,100)

plt.plot(x, funcA(x))
plt.xlabel("x")
plt.ylabel("y")
plt.title("Função A")
```
